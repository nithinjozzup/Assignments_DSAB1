# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Reading the dataset

```python
iris='iris.xls'
df=pd.read_excel(iris)
```

# 5.Visualizations

# 5.1 Box plot

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='SL',y='SW',hue='Classification',data=df)
plt.title('Sepal length vs Sepal width',size=20)
plt.figure(figsize=(20,10))
sns.boxplot(x='PL',y='PW',hue='Classification',data=df)
plt.title('Petal length vs Petal width',size=20)
```

# The box plot above gives a clear idea of how the sepal and petal length to width ratio changes according to the plant

#  5.2 Violin plot

```python
plt.figure(figsize=(15,15))
sns.violinplot(x='Classification',y='PL',data=df)
plt.title('Petal length of different Classification')
```

# 5.3 Strip plot

```python
plt.figure(figsize=(15,15))
sns.stripplot(x='Classification',y='SL',data=df)
plt.title('Sepal length of different Classification')
```

# 5.4 Swarm plot

```python
plt.figure(figsize=(15,15))
sns.swarmplot(x='Classification',y='SW',data=df,hue='Classification')
plt.title('Sepal width of different Classification')
```

# 5.5 Scatter plot

```python
plt.style.use('seaborn-deep')
plt.figure(figsize=(15,15))
sns.scatterplot(data=df,x ='PL', y="PW", hue='Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length and Width of Iris species')
```

# 5.6 Dis plot

```python
plt.figure(figsize=(15,15))
plt.style.use('Solarize_Light2')
sns.displot(data=df,x ='SW',hue='Classification',height=6.5,multiple='stack')
plt.title('Sepal Width count of Iris species')
```
